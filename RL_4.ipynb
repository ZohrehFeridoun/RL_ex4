```python
import gym
import numpy as np
import random

env = gym.make('FrozenLake8x8-v1')
```

```python
q_table = np.zeros((env.observation_space.n, env.action_space.n))
```

```python
def reward(state, action, next_state):
    if next_state == env.observation_space.n - 1:
        return 1
    else:
        return 0
```

```python
def sarsa(env, num_episodes, alpha, gamma, epsilon):
    for i in range(num_episodes):
        state = env.reset()
        action = get_action(state, epsilon, q_table)
        done = False
        while not done:
            next_state, reward_i, done, _ = env.step(action)
            next_action = get_action(next_state, epsilon, q_table)
            td_target = reward_i + gamma * q_table[next_state][next_action]
            td_error = td_target - q_table[state][action]
            q_table[state][action] += alpha * td_error
            state = next_state
            action = next_action
            reward_total = reward_i
        print(f"Episode {i+1}: Reward = {reward_total}")
```

```python
def get_action(state, epsilon, q_table):
    if random.uniform(0, 1) < epsilon:
        action = env.action_space.sample()
    else:
        action = np.argmax(q_table[state])
    return action
```

```python
num_episodes = 10000
alpha = 0.1
gamma = 0.99
epsilon = 0.05

sarsa(env, num_episodes, alpha, gamma, epsilon)
```
